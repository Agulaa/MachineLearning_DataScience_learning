```python
import pandas as pd 
import numpy as np 
import urllib 
```

```python
url = 'http://aima.cs.berkeley.edu/data/iris.csv'
req = urllib.request.Request(url)
data = urllib.request.urlopen(req)
names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'target']
iris_data = pd.read_csv(data, sep=',', decimal='.', header=None, names=names)
iris_data.head()
```

```python
#save iris data to cvs 
```

```python
iris_data.to_csv('iris_data.csv', sep=',', decimal='.', header=None)
```

```python
#Loading data in chunks 
```

```python
iris_chunks = pd.read_csv('iris_data.csv', header=None, names=names, chunksize=10)

for chunk in iris_chunks:
    print('Shape:', chunk.shape)
    print(chunk, '\n')
```

```python
iris_chunks_it = pd.read_csv('iris_data.csv', header=None, names=names, chunksize=10, iterator=True)
pices = iris_chunks_it.get_chunk(2)
pices
```
